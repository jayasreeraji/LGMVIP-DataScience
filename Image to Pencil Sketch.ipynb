# Image to Pencil Sketch with Python

```python
#importing require libraries
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#Reading the image
```

```python
img = cv2.imread("C:/Users/HP/OneDrive/Desktop/mandala reference/images (2).jpg")
```

```python
#Image open in new window
```

```python
cv2.imshow('Original image' ,img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

```python
#We will create a new image by converting the original image to greyscale
```

```python
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#The next step is to invert the new grayscale image
```

```python
invert_img = cv2.bitwise_not(grey_img)
```

```python
#Next step in the process is to blur the image by using the Gaussian Function in OpenCV
```

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
```

```python
invblur_img=cv2.bitwise_not(blur_img)
```

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
```

```python
#Saving the pencil sketch image
```

```python
cv2.imwrite('C:/Users/HP/OneDrive/Desktop/mandala reference/images (21).jpg', sketch_img)
```

```python
cv2.imshow('Sketch Image',sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#If you want to have a look at both the original image and the pencil sketch then you can use the following commands
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original Image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch Image', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
